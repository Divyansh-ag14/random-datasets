```python
sns.pairplot(df, hue="gender")
```

```python
cols = ["math score", 'reading score', "writing score"]
```

```python
for col in cols:
    sns.barplot("gender", col, data=df)
    plt.title(col + " vs Gender")
    plt.yticks(range(10,101,10))
    plt.show()
```

```python
for col in cols:
    sns.scatterplot("gender", col, data=df)
    plt.title(col + " vs Gender")
    plt.yticks(range(10,101,10))
    plt.show()
```

```python
for col in cols:
    sns.boxplot("gender", col, data=df)
    plt.title(col + " vs Gender")
    plt.yticks(range(10,101,10))
    plt.show()
```

```python
sns.scatterplot("reading score", "writing score", data=df, hue="gender")
```
